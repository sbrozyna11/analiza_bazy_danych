```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('~/sbrozyna/analiza_bazy_danych/TIER/original_data/billboard.csv')
print(type(df))
df.head()
```

```python
def uninvert(column):
    final_artists = []
    for record in column:
        artist = str(record)
        if ',' in artist:
            artist_bk = artist.split(',')
            artist = artist_bk[1].strip() + ' ' + artist_bk[0]
        final_artists.append(artist)
        print(final_artists)
    return final_artists
    
```

```python
inverted_artists = df["artist.inverted"].tolist()
uninverted_artists = uninvert(inverted_artists)
df["artist.inverted"] = uninverted_artists
df.head()
```
